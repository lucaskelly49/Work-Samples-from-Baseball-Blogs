```python
# importing new external libraries
import pandas as pd
```

```python
# grab the top 20 players from the wire, sorted by whatever 
# category you want, or whatever the default is

waiver = ['Matt Chapman',
'Yasmani Grandal',
"Tyler O'Neill",
'David Peralta',
'Michael Conforto',
'Josh Donaldson',
'Franmil Reyes',
'Kyle Seager',
'Gavin Lux',
'Pavin Smith',
'Anthony Santander',
'Jeff McNeil',
'Kyle Schwarber',
'Dominic Smith',
'Max Kepler',
'Marcell Ozuna',
'Jorge Polanco',
'Didi Gregorius',
'Kyle Lewis',
'David Fletcher']

# Players on my roster that I want to analyze

my_team = ['Yoan Moncada',
'Francisco Lindor',
'Eugenio Suarez',
'Brandon Lowe', 
'Fernando Tatis Jr.',
'J.T. Realmuto',
'Anthony Rizzo', 
'Jose Altuve', 
'Bryce Harper',
'Joey Gallo', 
'Ryan Mountcastle', 
'Wil Myers', 
'Ian Happ', 
'Giancarlo Stanton']
```

```python
# downloaded .csv from FanGraphs (limited to at least 150 PA)
df = pd.read_csv('xwOBARep.csv')

# using the lists I made above, creating two separate data sets
# and marking a column to indicate whether they are rostered or not
smores = df.loc[df['Name'].isin(my_team)][['Name', 'xwOBA']] #smores is my team name
smores['flag']=1

waivers = df.loc[df['Name'].isin(waiver)][['Name', 'xwOBA']]
waivers['flag']=0

# combine the two data frames into one
frames = [smores, waivers]
df2 = pd.concat(frames)

# sort based on xwOBA
df_final = df2.sort_values(by='xwOBA', ascending=False)
df_final
```
